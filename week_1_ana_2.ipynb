**Disorder Prevalence by Age Band**

 grouped ages into categories:

•	<20, 20–29, 30–39, 40–49, 50–59, 60–69, 70+

Then calculated:

•	Average Anxiety Score

•	Average Depression Score


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/anxiety_depression_data.csv')

# Create Age Bands
bins = [0, 20, 30, 40, 50, 60, 70, 100]
labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70+']
df['Age_Band'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Group data
grouped = df.groupby('Age_Band')[['Anxiety_Score', 'Depression_Score']].mean()

# Plot chart
plt.figure()
grouped.plot(kind='bar')

plt.xlabel('Age Band')
plt.ylabel('Average Score')
plt.title('Disorder Prevalence by Age Band')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/anxiety_depression_data.csv')

# Group by Gender
gender_group = df.groupby('Gender')[['Anxiety_Score', 'Depression_Score']].mean()

# Plot chart
plt.figure()
gender_group.plot(kind='bar')

plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.title('Disorder Prevalence by Gender')
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

**Age band vs Gender**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/anxiety_depression_data.csv')

# Create Age Bands
bins = [0, 20, 30, 40, 50, 60, 70, 100]
labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70+']
df['Age_Band'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Group by Age Band and Gender
grouped = df.groupby(['Age_Band', 'Gender'])[['Anxiety_Score', 'Depression_Score']].mean()

# Separate Anxiety and Depression for plotting
anxiety = grouped['Anxiety_Score'].unstack()
depression = grouped['Depression_Score'].unstack()

# Plot Anxiety chart
plt.figure()
anxiety.plot(kind='bar')
plt.title('Age Band vs Gender (Anxiety Score)')
plt.xlabel('Age Band')
plt.ylabel('Average Anxiety Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot Depression chart
plt.figure()
depression.plot(kind='bar')
plt.title('Age Band vs Gender (Depression Score)')
plt.xlabel('Age Band')
plt.ylabel('Average Depression Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
