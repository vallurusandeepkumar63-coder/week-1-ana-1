```python

import pandas as pd
import plotly.express as px
```

```python

mental_health_df = pd.read_csv('/content/anxiety_depression_data.csv')
regional_df = pd.read_csv('/content/regional_prevalence.csv')

print("Mental Health Dataset Shape:", mental_health_df.shape)
print("Regional Dataset Shape:", regional_df.shape)
```

**Choropleth Map **

```python

# Choropleth Map using Plotly

fig = px.choropleth(
    regional_df,
    locations='Region',
    locationmode='country names',
    color='Prevalence (%)',
    hover_name='Region',
    color_continuous_scale='Viridis',
    title='Mental Health Disorder Prevalence by UK Region'
)

fig.update_layout(
    width=1000,
    height=700
)

fig.show()
```

```python

fig_bar = px.bar(
    regional_df,
    x='Region',
    y='Prevalence (%)',
    color='Prevalence (%)',
    title='Regional Mental Health Disorder Prevalence'
)

fig_bar.update_layout(
    xaxis_title='UK Region',
    yaxis_title='Prevalence (%)',
    width=1000,
    height=600
)

fig_bar.show()
```
